# Ejercicio#1

Calcular el valor del Seno de x con el uso de la siguiente sumatoria, considerar una tolerancia de 0.05%.

Calcular el valor real del seno con la función math.sin(x)  
Calcular el factorial con con la función math.factorial(x)  


$$sin(x)=\sum_{k=0}^{\infty}\frac{(-1)^{k}x^{2k+1}}{(2k+1)!}$$
$$error=\left |  \frac{valorReal - aproximacion}{valorReal}\right |*100$$

###Caso de prueba
x = 0.5  
Tolerancia = 0.05%  
Verificar que la **aproximación** sea un valor cercano al **valor real**  
Y que el **error** sea menor a la **tolerancia**

```
import math
x = float(input("x="))
tolerancia = float(input("Tolerancia="))
valorReal = math.sin(x)
k = 0
aprox = 0
while True:
  aprox += ((-1)**k*x**(2*k+1))/math.factorial(2*k+1)
  k += 1
  error = math.fabs((valorReal - aprox)/valorReal) * 100
  if error < tolerancia:
    break

print("Valor real = ", valorReal)
print("Aproximación = ", aprox)
print("Error = ",error)
```

#Ejercicio #4
Un gusano está en el fondo de un pozo de n pulgadas de profundidad. Tiene energía suficiente para trepar u pulgadas por minuto, pero luego tiene que descansar un minuto antes de subir de nuevo. Durante el descanso, se desliza hacia abajo d pulgadas. Este proceso de escalada y de descanso, se repite. ¿Cuánto tiempo pasara antes de que el gusano logre salir del pozo? Siempre se cuenta una parte de minuto, como un minuto y si el gusano acaba de llegar a la cima al final de su escalada, vamos a asumir el gusano ya esta fuera.

###Caso de prueba
n = 10    
u = 2  
d = 1    
Tiempo: 17  

n = 20    
u = 3  
d = 1    
Tiempo: 19  


```
n=int(input("Profundidad del pozo(pulg)="))
u=int(input("Energía(pulg/min)="))
d = 1
ascenso=0
tiempo  = 0
while True:
  ascenso += u #desplazamiento hacia arriba
  tiempo += 1  #Tiempo de ascenso
  if ascenso >= n:  #Si llegó a la cima sale del ciclo
    break;
  ascenso -= d #Desplazamiento hacia abajo
  tiempo += 1  #Tiempo de descanso

print("Tiempo en salir:",tiempo)
```
