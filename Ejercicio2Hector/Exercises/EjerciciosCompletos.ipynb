# Ejercicio#1

Calcular el valor del Seno de x con el uso de la siguiente sumatoria, considerar una tolerancia de 0.05%.

Calcular el valor real del seno con la función math.sin(x)  
Calcular el factorial con con la función math.factorial(x)  


$$sin(x)=\sum_{k=0}^{\infty}\frac{(-1)^{k}x^{2k+1}}{(2k+1)!}$$
$$error=\left |  \frac{valorReal - aproximacion}{valorReal}\right |*100$$

###Caso de prueba
x = 0.5  
Tolerancia = 0.05%  
Verificar que la **aproximación** sea un valor cercano al **valor real**  
Y que el **error** sea menor a la **tolerancia**

```python
#Fomra original sin funciones
import math
x = float(input("x="))
tolerancia = float(input("Tolerancia="))
valorReal = math.sin(x)
k = 0
aprox = 0
while True:
  aprox += ((-1)**k*x**(2*k+1))/math.factorial(2*k+1)
  k += 1
  error = math.fabs((valorReal - aprox)/valorReal) * 100
  if error < tolerancia:
    break

print("Valor real = ", valorReal)
print("Aproximación = ", aprox)
print("Error = ",error)
```
